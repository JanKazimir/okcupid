### Splitting the data set into waves

```python
wave1_part = []
wave2_part = []
wave3_part = []
others_part = []
for column in df.columns:
    if column.startswith("w1") | column.startswith("p17"):
        wave1_part.append(column)
    elif column.startswith("w2")| column.startswith("p18")| column.startswith("p19")| column.startswith("p20"):
        wave2_part.append(column)
    elif column.startswith("w3"):
        wave3_part.append(column)
    else:
        others_part.append(column)

df_w1 = df[wave1_part+others_part]
df_w2 = df[wave2_part+others_part]
df_w3 = df[wave3_part+others_part]

```

```python
# Define the subsets based on codebook definitions
dating_col = 'w3_q24_internet_dating'
other_cols = [
    'w3_q24_internet_soc_network', 
    'w3_q24_internet_game', 
    'w3_q24_internet_chat', 
    'w3_q24_internet_org'
]

# 1. Create 'internet_dating' (Ensure it is binary 0/1)
df['internet_dating'] = df[dating_col].fillna(0).astype(int)

# 2. Create 'other_internet_methods' 
# Logic: If any of the 'other_cols' are 1, the result is 1.
df['other_internet_methods'] = df[other_cols].max(axis=1).fillna(0).astype(int)

# 3. Final reduction: Keep only the two desired columns (plus Case ID for merging)
df_final = df[['caseid_new', 'internet_dating', 'other_internet_methods']]

df_final
```
