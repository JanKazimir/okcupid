```python
wave1_part = []
wave2_part = []
wave3_part = []
others_part = []
for column in df.columns:
    if column.startswith("w1") | column.startswith("p17"):
        wave1_part.append(column)
    elif (
        column.startswith("w2")
        | column.startswith("p18")
        | column.startswith("p19")
        | column.startswith("p20")
    ):
        wave2_part.append(column)
    elif column.startswith("w3"):
        wave3_part.append(column)
    else:
        others_part.append(column)
```

```python
df_w1 = df[wave1_part + others_part]
df_w2 = df[wave2_part + others_part]
df_w3 = df[wave3_part + others_part]
```

```python
categorical_object_columns = []
binary_columns = []
categorical_columns = []

for column in df_w1.columns:
    if str(df_w1[column].values).__contains__("Categories") and str(
        df_w1[column].values
    ).__contains__(", str)"):
        if str(df_w1[column].values).__contains__("(2, str)") and (
            str(df_w1[column].values).__contains__("Yes")
            or str(df_w1[column].values).__contains__("yes")
        ):
            binary_columns.append({column: df_w1[column].unique().tolist()})
        else:
            categorical_columns.append({column: df_w1[column].unique().tolist()})

```

```python
single_value_columns = []
binary_columns = []
categorical_columns = []


# 1. Get all category and object columns
cat_cols = df_w1.select_dtypes(include=["category"]).columns

# 2. Process Categorical columns
for col in cat_cols:
    # nunique(dropna=True) ignores NaN when counting
    unique_count = df_w1[col].nunique(dropna=True)
    unique_vals = df_w1[col].unique().tolist()

    if unique_count == 1:
        single_value_columns.append({col: unique_vals})
    elif unique_count == 2:
        binary_columns.append({col: unique_vals})
    else:
        categorical_columns.append({col: unique_vals})

# Create the yes_no list
yes_no_columns = [
    col
    for col in binary_columns
    if any("yes" in str(val).lower() for val in list(col.values())[0])
]

# Update binary_columns to only keep those NOT in yes_no_columns
binary_columns = [col for col in binary_columns if col not in yes_no_columns]
```

```python
from collections import defaultdict

shared_sets = defaultdict(list)

for col_dict in categorical_columns:
    for col_name, values in col_dict.items():
        # Using sorted tuple as key to maintain value consistency
        key = tuple(sorted([str(v) for v in values]))
        shared_sets[key].append(col_name)

# Display only the sets shared by multiple columns
for values, cols in shared_sets.items():
    if len(cols) > 1:
        print(f"Columns: {cols}\nValues: {values}\n")
```

```python
import numpy as np

# 1. Define your mapping logic
# This handles variations in capitalization and whitespace
yes_no_refused_map = {"yes": 1, "no": 0, "refused": -1,  "nan": -1}

# 2. Iterate through your shared_sets to find the target columns
for values_tuple, col_list in shared_sets.items():
    # Check if the set contains 'Yes' and 'No'
    # This targets groups like ('No', 'Refused', 'Yes') and ('No', 'Refused', 'Yes', 'nan')
    vals_lower = [str(v).lower() for v in values_tuple]

    if "yes" in vals_lower and "no" in vals_lower:
        for col in col_list:
            # Apply the mapping
            # .map() is much faster than a loop for this
            df_w1[col] = (
                df_w1[col].astype(str).str.lower().str.strip().map(yes_no_refused_map)
            )

            # Optional: fill any remaining unmapped values with -1
            df_w1[col] = df_w1[col].fillna(-1).astype(int)

print("Standardization complete for Yes/No/Refused columns.")
```
